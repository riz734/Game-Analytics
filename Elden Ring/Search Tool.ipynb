```python
import pandas as pd

from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import os

#This function takes in a text string then returns lst of tuples of top results using token sort ration
#It looks through all datasets for matching or similar results
def search(search_string):

    arr = os.listdir("data")
    
    #stores search results into diction
    top_results = []
    
    #loops through list of loaded file names
    for i in arr:
        
        if "csv" in i:
            
            #loads each datasets
            df = pd.read_csv("data/"+i)
            
            #loops through name column in each datasets
            for i in df["name"]:
                
                #checks matching of values in name column
                if fuzz.token_sort_ratio(search_string, i) > 50:
                    
                    #filters to the row where the search is similar
                    result = df[df["name"]==i]
                    
                    #appends the result to top_results, list of tuples where [0] is the df and [1] is token ratio
                    
                    top_results.append((result,fuzz.token_sort_ratio(search_string, i)))
    
    #sorts on token ratio descending
    sorted_results = sorted(top_results,key=lambda x: x[1], reverse=True)
                
    return sorted_results
```

```python
#top results of "giant"
search("giant")[0]
```
